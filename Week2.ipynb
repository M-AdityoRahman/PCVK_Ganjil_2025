1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?

OpenCV (cv2_imshow) menampilkan gambar dalam format BGR.

Matplotlib (plt.imshow) menampilkan dalam format RGB. Konversi diperlukan untuk menampilkan warna yang benar.

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?

int16 memiliki range -32768 hingga 32767.

int32 memiliki range yang lebih besar.

Untuk gambar, uint8 (0-255) lebih umum digunakan

7. Buat rectangle dan circle pada bagian wajah dari image foto anda saat
beraktifitas (bukan pasfoto).

```python
# Ganti dengan path foto Anda
img_self = cv.imread('/content/drive/MyDrive/PCVK/oasis.jpeg')
img_self_rgb = cv.cvtColor(img_self, cv.COLOR_BGR2RGB)

# Contoh koordinat wajah (sesuaikan dengan foto Anda)
x, y, w, h = 100, 100, 200, 200  # Ganti dengan koordinat wajah

# Rectangle
cv.rectangle(img_self_rgb, (x, y), (x+w, y+h), (255,0,0), 5)

# Circle
center = (x + w//2, y + h//2)
radius = w//2
cv.circle(img_self_rgb, center, radius, (0,255,0), 5)

plt.imshow(img_self_rgb)
plt.show()
```
